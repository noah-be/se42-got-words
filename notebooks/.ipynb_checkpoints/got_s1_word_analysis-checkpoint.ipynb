```python
freq_values = freq_df["count"].values
ranks = np.arange(1, len(freq_values) + 1)

plt.figure(figsize=(8,6))
plt.loglog(ranks, freq_values, marker=".", linestyle="none")
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.title("Zipf's Law in Game of Thrones Season 1")
plt.tight_layout()
plt.show()
```

```python
words_per_episode = (
    df.groupby("episode")["tokens"]
    .apply(lambda col: sum(len(tokens) for tokens in col))
    .reset_index(name="word_count")
)

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(words_per_episode["episode"], words_per_episode["word_count"])
plt.xlabel("Episode")
plt.ylabel("Number of word tokens")
plt.title("Total dialogue volume per episode")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
